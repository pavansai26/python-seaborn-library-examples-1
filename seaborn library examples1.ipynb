```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='line')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',markers='+')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,markers='+',hue='sex',size='size')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,markers='+',hue='sex',size='size',palette='ch:0.9,0.8')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,markers='+',hue='sex',size='size',sizes=(10,500))
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x='time',y='value',data=df,kind='line')
```

```python
sns.relplot(x='time',y='value',data=df,kind='line',sort=False,ci=False)
```

```python
fmri=sns.load_dataset('fmri')
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',ci=False)
```

```python
palette=sns.cubehelix_palette(light=0.8,n_colors=6)
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',ci=False,palette=palette)
```

```python
dots=sns.load_dataset('dots')
palette=sns.cubehelix_palette(light=0.8,n_colors=6)
sns.relplot(x='time',y='firing_rate',data=dots,kind='scatter',ci=False,palette=palette,sizes=(1,5))
```
